# 3. 여러 페이지 HTML 갖고 오기 

### 3.1 컬리

```python
kurly = 'https://www.kurly.com/'

# print(kurly)
# print(kurly.text)

for i in range(1,4):
    # https://www.kurly.com/search?sword=%ED%81%AC%EB%A6%AC%EC%8A%A4%EB%A7%88%EC%8A%A4&page=3
    kurly_page = requests.get(f'{kurly}/categories/907?page={i}')
    print(f'page{i}: {kurly_page.text}')
```

```python
kurly_pages = []

for i in range(1, 4):
    url = f'https://www.kurly.com/search?sword=%ED%81%AC%EB%A6%AC%EC%8A%A4%EB%A7%88%EC%8A%A4&page={i}'

    response = requests.get(url)
    print(response.text)

    kurly_pages.append(response.text)

print(len(kurly_pages))

for i in range(len(kurly_pages)):
    print(kurly_pages[i])
    print()
    
```

### 3.2 네이버 쇼핑 

```python
naver_shopping = 'https://shopping.naver.com/'

# print(naver_shopping)
# print(naver_shopping.text)

for i in range(1,4):
    # https://search.shopping.naver.com/ns/search?query=%ED%81%AC%EB%A6%AC%EC%8A%A4%EB%A7%88%EC%8A%A4&listPage=9
    naver_shopping_page = requests.get(f'{naver_shopping}/ns/search?query=크리스마스&listPage{i}')
    print(f'page{i}: {naver_shopping_page.text}')
```

### 3.3 옥션

```python
auction = 'https://www.auction.co.kr'

# print(auction)
# print(auction.text)

for i in range(1,4):
    # https://www.auction.co.kr/n/search?keyword=%ed%81%ac%eb%a6%ac%ec%8a%a4%eb%a7%88%ec%8a%a4&k=52&p=3
    auction_page = requests.get(f'{auction}/n/search?keyword=크리스마스&k=52&p={i}')
    print(f'page{i}: {auction_page.text}')
```
