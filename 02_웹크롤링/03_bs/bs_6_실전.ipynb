```python
import requests 
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/sise/sise_quant.nhn'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


tr_tags = soup.select('table.type_2 tr')
len(tr_tags)

rising_stock = []

for tag in tr_tags: # 각 tr태그 갖고옴 
    # print(len(tag.select('td'))) # tr마다 td가 몇개인지 확인 

    if len(tag.select('td')) != 12: # td가 12개가 아니면 for문 건너뛰기 
        continue 

    # td가 12개라면 
    if tag.select('td')[3].find('span').text.strip() != '상승':
        continue

    # span.text가 상승이면 
    rank = tag.select('td')[0].text.strip()
    name = tag.select('td')[1].text.strip()
    presentPrice = tag.select('td')[2].text.strip()
    yesterdaty = tag.select('td')[3].select('span')[1].text.strip()
    # print(yesterdaty)

    rising_stock.append(dict(순위=rank, 종목=name, 현재가=presentPrice, 전일비=yesterdaty))


    # break

print(len(rising_stock))
rising_stock
```

## 4.1 다음 뉴스 

```python
import requests
from bs4 import BeautifulSoup

daum_url = 'https://news.daum.net/breakingnews/society?page=1'
# url = 'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date=20240803&page=1'

response = requests.get(daum_url)
soup = BeautifulSoup(response.text, 'html.parser')

li_tags = soup.select('ul.list_news2.list_allnews li')
for tag in li_tags:
    # print(tag.select('dt')[1].select_one('a').attrs['href'])
    news_url = ''
    
    try:
        news_url = tag.select_one('strong.tit_thumb a.link_txt').attrs['href']
    except:
        news_url = tag.select_one('a.link_thumb').attrs['href']


    # print(news_url)
    response = requests.get(news_url)
    news_soup = BeautifulSoup(response.text, 'html.parser')

    # title = news_soup.select_one('#title_area span').text.strip()
    title = news_soup.select_one('div.head_view h3.tit_view').text.strip()
    body_tags = news_soup.select('div.article_view section p')
    body = []
    for p_tag in body_tags:
        body.append(p_tag.text)


    print(title)
    print(''.join(body))
    print()
```

## 4.2 네이버 뉴스 

- 한 페이지에서 데이터 뽑기 

```python
import requests
from bs4 import BeautifulSoup 

naverNews_src = requests.get('https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032')
soup = BeautifulSoup(naverNews_src.text, 'html.parser')

ul_tags = soup.select('div.list_body.newsflash_body ul')
all_datas = []

for ul_tag in ul_tags:

    li_tags = ul_tag.select('li')

    for li_tag in li_tags:

        try:
            url = li_tag.select('a')[0].attrs['href']
        except:
            url = li_tag.select('a')[1].attrs['href']

        # print(url)
        # break

        news_res = requests.get(url)
        news_soup = BeautifulSoup(news_res.text, 'html.parser')

        title = news_soup.select_one('#title_area span').text

        body = news_soup.select_one('article').find_all(string=True, recursive=False)
        body = ''.join(body).strip()
        

        all_datas.append(dict(
            title = title,
            body = body
        ))

print(len(all_datas))
print(all_datas[:5])
```
