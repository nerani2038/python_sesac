아래 사이트를 크롤링하여 아래와 같이 각각 글에 id와 title 그리고 글마다 코멘트내용을 리스트형식으로 담고 최종 json 파일 형태로 저장해보세요.


```python
url = 'https://crawlingstudy-dd3c9.web.app/05/'
photo_url = 'https://jsonplaceholder.typicode.com/photos'
comment_url = 'https://jsonplaceholder.typicode.com/comments?postId=1'
```

```python
import requests
import json
photo_response = requests.get(photo_url)
photo_json = photo_response.json()
# print(photo_json)
result = []
for photo in photo_json[:10]:
    id = photo['id']
    title = photo['title']

    comment_url = 'https://jsonplaceholder.typicode.com/comments?postId=' + str(id)
    comment_response = requests.get(comment_url)
    comment_json = comment_response.json()

    com_list = []
    for comment in comment_json:
        com = comment['body']
        com_list.append(com)
    
    result.append(dict(
        id=id,
        title=title,
        comment=com_list
    ))

# 10개만
# id
# title
# comment
# list 화
# json 형식으로 저장 - 파일로 저장

# data = json.dumps(result)
with open('comment_prac.json', 'w') as json_file:
    json.dump(result, json_file)



```

```python
# 쌤답안

# 비동기 실습 1번 답안코드 입니다~

import requests
import json

# api 서버에 데이터 요청하기 
api_response = requests.get('https://jsonplaceholder.typicode.com/photos')

title_json = json.loads(api_response.text)


result = []
for data in title_json[:10]: 

    temp = dict(
        id=data['id'], 
        title=data['title'], 
        comments=[]
    )

    comment_res = requests.get(f'https://jsonplaceholder.typicode.com/comments?postId={data["id"]}')
    for comment in comment_res.json():
        temp['comments'].append(comment['body'])

    result.append(temp)
    
    # break # 한 싸이클 확인용  (다 하면 5000개 넘음..)


# 저장
with open('asynch_data.json', 'w') as json_file:
    json.dump(result, json_file) # datas를 json_file에 json포맷으로 작성후 저장 
```

# 둘의 차이점?

| **구분**          | **`api_response.json()`**          | **`json.loads(api_response.text)`**  |
|--------------------|------------------------------------|---------------------------------------|
| **소속**          | `requests.Response`의 메서드      | `json` 모듈의 함수                   |
| **입력값**        | 내부적으로 처리된 JSON 응답       | 명시적으로 문자열을 입력             |
| **동작 방식**     | 자동으로 JSON으로 파싱            | 문자열을 JSON으로 변환               |
| **코드 간결성**   | 더 간결함                         | 비교적 덜 간결함                     |
| **에러 처리**     | JSON이 아니면 `ValueError` 발생   | JSON이 아니면 `ValueError` 발생      |
| **사용 목적**     | `requests` 응답에서 바로 사용      | 문자열 데이터를 명시적으로 변환       |

```python
json_url = 'https://jsonplaceholder.typicode.com/photos'

response = requests.get(json_url)
# print(response.text)

# 파서는 포맷마다 존재함 - 내가 어떤형태의 데이터를 받냐에 따라 다른 파서를 사용해야함

# response가 html일 경우의 파싱 방식
soup = BeautifulSoup(response.text, 'html.parser')

# json 포맷으로 구조화해주는 파싱 방식
title_data = response.json() # .json -> json 포맷으로 구조화 시켜줄 파서가 됨

```

```python
comment_url = 'https://jsonplaceholder.typicode.com/comments?postId=1'

comment_response = requests.get(comment_url) # 아직까지는 str 형식
comment_data = comment_response.json()

# print(type(comment_data))
# print(type(comment_data[0]))
# comment_data
```

```python
print(len(title_data))
print(len(comment_data)) # 그냥 값을 가져오는것이 아니라 하나씩 확인하면서 동작시켜야함
```

```python
result = []

for data in title_data[:10]:

    id = data['id']
    title = data['title']

    temp = dict(
        id = id,
        title = title,
        comments = []
    )

    comment_url = f'https://jsonplaceholder.typicode.com/comments?postId={id}'
    res = requests.get(comment_url)
    comment_data = res.json()

    for comment in comment_data:
        temp['comments'].append(comment['body'])
    
    result.append(temp)

result
```

```python
import json

with open('async_data.json', 'w') as json_file:
    json.dump(result, json_file)

# 이렇게 따로 저장소에 json파일을 저장할수있다.
```
